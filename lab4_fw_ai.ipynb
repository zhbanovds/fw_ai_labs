## Выбор наборов данных

Для задачи классификации используется датасет Breast Cancer Wisconsin Diagnostic. Для задачи регрессии используется датасет Concrete Compressive Strength.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, f1_score, classification_report, confusion_matrix
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
```

## Загрузка данных


```python
columns = [
    "id", "diagnosis",
    "radius_mean", "texture_mean", "perimeter_mean", "area_mean", "smoothness_mean",
    "compactness_mean", "concavity_mean", "concave_points_mean", "symmetry_mean", "fractal_dimension_mean",
    "radius_se", "texture_se", "perimeter_se", "area_se", "smoothness_se",
    "compactness_se", "concavity_se", "concave_points_se", "symmetry_se", "fractal_dimension_se",
    "radius_worst", "texture_worst", "perimeter_worst", "area_worst", "smoothness_worst",
    "compactness_worst", "concavity_worst", "concave_points_worst", "symmetry_worst", "fractal_dimension_worst"
]

df_cls = pd.read_csv("data/breast_cancer/wdbc.data", header=None, names=columns)
df_reg = pd.read_excel("data/concrete/concrete.xls")
```

## Выбор целевых переменных и метрик качества

Для классификации используются accuracy и F1-score. Для регрессии используются MAE, MSE и R2.


```python
X_cls = df_cls.drop(columns=["id", "diagnosis"])
y_cls = df_cls["diagnosis"]

encoder = LabelEncoder()
y_cls = encoder.fit_transform(y_cls)

X_reg = df_reg.iloc[:, :-1]
y_reg = df_reg.iloc[:, -1]

print(X_cls.shape, len(y_cls))
print(X_reg.shape, len(y_reg))
print(encoder.classes_)
```

## Разделение данных на обучающую и тестовую выборки


```python
X_train_cls, X_test_cls, y_train_cls, y_test_cls = train_test_split(
    X_cls, y_cls, test_size=0.2, random_state=42, stratify=y_cls
)

X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_reg, y_reg, test_size=0.2, random_state=42
)
```

## Построение базового решения на sklearn

На данном этапе строятся базовые модели случайного леса для классификации и регрессии.


```python
rf_cls_base = RandomForestClassifier(random_state=42)
rf_cls_base.fit(X_train_cls, y_train_cls)
y_pred_cls_base = rf_cls_base.predict(X_test_cls)

acc_base = accuracy_score(y_test_cls, y_pred_cls_base)
f1_base = f1_score(y_test_cls, y_pred_cls_base)

print("Baseline classification accuracy:", acc_base)
print("Baseline classification f1:", f1_base)
```

```python
print(classification_report(y_test_cls, y_pred_cls_base))
print(confusion_matrix(y_test_cls, y_pred_cls_base))
```

```python
rf_reg_base = RandomForestRegressor(random_state=42)
rf_reg_base.fit(X_train_reg, y_train_reg)
y_pred_reg_base = rf_reg_base.predict(X_test_reg)

mae_base = mean_absolute_error(y_test_reg, y_pred_reg_base)
mse_base = mean_squared_error(y_test_reg, y_pred_reg_base)
r2_base = r2_score(y_test_reg, y_pred_reg_base)

print("Baseline regression MAE:", mae_base)
print("Baseline regression MSE:", mse_base)
print("Baseline regression R2:", r2_base)
```

## Формулировка гипотез улучшения

1. Количество деревьев в ансамбле влияет на качество модели.
2. Ограничение глубины деревьев может снизить переобучение.
3. Параметры min_samples_split и min_samples_leaf влияют на качество модели.
4. Выбор числа признаков, рассматриваемых при каждом разбиении, влияет на качество.
5. Подбор гиперпараметров на кросс-валидации позволяет получить более качественное решение.


## Проверка гипотезы о количестве деревьев


```python
n_estimators_values = [10, 25, 50, 100, 150, 200]

cls_estimators_scores = []
for n in n_estimators_values:
    model = RandomForestClassifier(n_estimators=n, random_state=42)
    model.fit(X_train_cls, y_train_cls)
    pred = model.predict(X_test_cls)
    cls_estimators_scores.append(f1_score(y_test_cls, pred))

reg_estimators_scores = []
for n in n_estimators_values:
    model = RandomForestRegressor(n_estimators=n, random_state=42)
    model.fit(X_train_reg, y_train_reg)
    pred = model.predict(X_test_reg)
    reg_estimators_scores.append(r2_score(y_test_reg, pred))
```

```python
plt.figure(figsize=(8, 4))
plt.plot(n_estimators_values, cls_estimators_scores, marker='o')
plt.xlabel("n_estimators")
plt.ylabel("F1-score")
plt.title("Random Forest Classification Quality for Different n_estimators")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.plot(n_estimators_values, reg_estimators_scores, marker='o')
plt.xlabel("n_estimators")
plt.ylabel("R2")
plt.title("Random Forest Regression Quality for Different n_estimators")
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Проверка гипотезы о глубине деревьев


```python
depth_values = [2, 3, 4, 5, 6, 8, 10, None]

cls_depth_scores = []
for depth in depth_values:
    model = RandomForestClassifier(n_estimators=100, max_depth=depth, random_state=42)
    model.fit(X_train_cls, y_train_cls)
    pred = model.predict(X_test_cls)
    cls_depth_scores.append(f1_score(y_test_cls, pred))

reg_depth_scores = []
for depth in depth_values:
    model = RandomForestRegressor(n_estimators=100, max_depth=depth, random_state=42)
    model.fit(X_train_reg, y_train_reg)
    pred = model.predict(X_test_reg)
    reg_depth_scores.append(r2_score(y_test_reg, pred))
```

```python
depth_labels = [str(x) for x in depth_values]

plt.figure(figsize=(8, 4))
plt.plot(depth_labels, cls_depth_scores, marker='o')
plt.xlabel("max_depth")
plt.ylabel("F1-score")
plt.title("Random Forest Classification Quality for Different Depth")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.plot(depth_labels, reg_depth_scores, marker='o')
plt.xlabel("max_depth")
plt.ylabel("R2")
plt.title("Random Forest Regression Quality for Different Depth")
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Подбор гиперпараметров на кросс-валидации


```python
param_grid_cls = {
    "n_estimators": [50, 100, 200],
    "max_depth": [3, 5, 8, None],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4],
    "max_features": ["sqrt", "log2", None]
}

grid_cls = GridSearchCV(
    RandomForestClassifier(random_state=42),
    param_grid_cls,
    cv=5,
    scoring="f1",
    n_jobs=-1
)

grid_cls.fit(X_train_cls, y_train_cls)

print(grid_cls.best_params_)
print(grid_cls.best_score_)
```

```python
param_grid_reg = {
    "n_estimators": [50, 100, 200],
    "max_depth": [3, 5, 8, None],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4],
    "max_features": ["sqrt", "log2", None]
}

grid_reg = GridSearchCV(
    RandomForestRegressor(random_state=42),
    param_grid_reg,
    cv=5,
    scoring="r2",
    n_jobs=-1
)

grid_reg.fit(X_train_reg, y_train_reg)

print(grid_reg.best_params_)
print(grid_reg.best_score_)
```

## Самостоятельная имплементация алгоритмов

Далее реализуются собственные версии случайного леса для классификации и регрессии на основе ансамбля решающих деревьев.


```python
class MyRandomForestClassifier:
    def __init__(self, n_estimators=10, max_depth=5, min_samples_split=2, max_features=None, random_state=42):
        self.n_estimators = n_estimators
        self.max_depth = max_depth
        self.min_samples_split = min_samples_split
        self.max_features = max_features
        self.random_state = random_state
        self.trees = []
        self.feature_subsets = []

    def fit(self, X, y):
        rng = np.random.default_rng(self.random_state)
        X = np.array(X)
        y = np.array(y)
        n_samples, n_features = X.shape

        if self.max_features is None:
            n_select = n_features
        elif self.max_features == "sqrt":
            n_select = max(1, int(np.sqrt(n_features)))
        elif self.max_features == "log2":
            n_select = max(1, int(np.log2(n_features)))
        else:
            n_select = int(self.max_features)

        self.trees = []
        self.feature_subsets = []

        for _ in range(self.n_estimators):
            sample_idx = rng.choice(n_samples, size=n_samples, replace=True)
            feature_idx = rng.choice(n_features, size=n_select, replace=False)

            X_sample = X[sample_idx][:, feature_idx]
            y_sample = y[sample_idx]

            tree = DecisionTreeClassifier(
                max_depth=self.max_depth,
                min_samples_split=self.min_samples_split,
                random_state=int(rng.integers(0, 1_000_000))
            )
            tree.fit(X_sample, y_sample)

            self.trees.append(tree)
            self.feature_subsets.append(feature_idx)

    def predict(self, X):
        X = np.array(X)
        predictions = []

        for tree, feature_idx in zip(self.trees, self.feature_subsets):
            predictions.append(tree.predict(X[:, feature_idx]))

        predictions = np.array(predictions)
        final_predictions = []

        for i in range(X.shape[0]):
            values, counts = np.unique(predictions[:, i], return_counts=True)
            final_predictions.append(values[np.argmax(counts)])

        return np.array(final_predictions)
```

```python
class MyRandomForestRegressor:
    def __init__(self, n_estimators=10, max_depth=5, min_samples_split=2, max_features=None, random_state=42):
        self.n_estimators = n_estimators
        self.max_depth = max_depth
        self.min_samples_split = min_samples_split
        self.max_features = max_features
        self.random_state = random_state
        self.trees = []
        self.feature_subsets = []

    def fit(self, X, y):
        rng = np.random.default_rng(self.random_state)
        X = np.array(X)
        y = np.array(y)
        n_samples, n_features = X.shape

        if self.max_features is None:
            n_select = n_features
        elif self.max_features == "sqrt":
            n_select = max(1, int(np.sqrt(n_features)))
        elif self.max_features == "log2":
            n_select = max(1, int(np.log2(n_features)))
        else:
            n_select = int(self.max_features)

        self.trees = []
        self.feature_subsets = []

        for _ in range(self.n_estimators):
            sample_idx = rng.choice(n_samples, size=n_samples, replace=True)
            feature_idx = rng.choice(n_features, size=n_select, replace=False)

            X_sample = X[sample_idx][:, feature_idx]
            y_sample = y[sample_idx]

            tree = DecisionTreeRegressor(
                max_depth=self.max_depth,
                min_samples_split=self.min_samples_split,
                random_state=int(rng.integers(0, 1_000_000))
            )
            tree.fit(X_sample, y_sample)

            self.trees.append(tree)
            self.feature_subsets.append(feature_idx)

    def predict(self, X):
        X = np.array(X)
        predictions = []

        for tree, feature_idx in zip(self.trees, self.feature_subsets):
            predictions.append(tree.predict(X[:, feature_idx]))

        predictions = np.array(predictions)
        return np.mean(predictions, axis=0)
```

## Обучение и оценка собственной реализации


```python
my_rf_cls = MyRandomForestClassifier(n_estimators=20, max_depth=5, min_samples_split=2, max_features="sqrt", random_state=42)
my_rf_cls.fit(X_train_cls, y_train_cls)
y_pred_my_cls = my_rf_cls.predict(X_test_cls)

my_acc = accuracy_score(y_test_cls, y_pred_my_cls)
my_f1 = f1_score(y_test_cls, y_pred_my_cls)

print("My random forest classification accuracy:", my_acc)
print("My random forest classification f1:", my_f1)
```

```python
my_rf_reg = MyRandomForestRegressor(n_estimators=20, max_depth=5, min_samples_split=2, max_features="sqrt", random_state=42)
my_rf_reg.fit(X_train_reg, y_train_reg)
y_pred_my_reg = my_rf_reg.predict(X_test_reg)

my_mae = mean_absolute_error(y_test_reg, y_pred_my_reg)
my_mse = mean_squared_error(y_test_reg, y_pred_my_reg)
my_r2 = r2_score(y_test_reg, y_pred_my_reg)

print("My random forest regression MAE:", my_mae)
print("My random forest regression MSE:", my_mse)
print("My random forest regression R2:", my_r2)
```

Затем были реализованы собственные версии алгоритма случайного леса для классификации и регрессии и проведено их сравнение с реализациями из sklearn.
